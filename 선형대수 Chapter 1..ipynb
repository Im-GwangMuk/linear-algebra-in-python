### 선형성(Linearity)

$f(x)$ 를 함수(function)또는 연산(operation)이라고 할때 다음 두가지 조건을 만족하면 선형성을 가진다.

- Superposition : $f(x_1+x_2) = f(x_1)+f(x_2)$


- homogeneity : $f(ax) = af(x)$ (여기서 a는 상수)


즉, $f(a_1 x_1+a_2x_2) = a_1f(x_1)+a_2f(x_2)$

여기서는 이런 선형성을 만족하는 것들을 다룰 것이다.

#### (3) 가우스 소거법(Gaussian Elimination)

가우스 소거법은 AX=B 라는 행렬에 대해 A를 상삼각행렬로 만들기 위해 수행해 주는 소거법이다. A의 형태를 상삼각행렬로 만들어 주게 되면 가장 아래에 있는 행의 변수는 한가지 밖에 없게되어서 쉽게 해를 구할 수 있다. 예를들어 설명하면 다음과 같다.

$$2u + v + w = 5 ...㉠ \\ 4u-6v=-2...㉡\\-2u+7v+2w=9...㉢$$

위와 같은 방정식 ㉠,㉡,㉢가 있을때 상삼각행렬로 만들어 주기 위해서 $㉡-2\times㉠$를 하고 $㉢+㉠$을 하면 다음과 같이 된다.

$$2u + v + w = 5 ...㉠ \\ -8v-2w=-12...㉡\\8v+3w=14...㉢$$

또, 이 때 ㉢+㉡을 해주면 다음과 같이 상삼각행렬을 만들어 준다.

$$2u + v + w = 5...㉠ \\ -8v -2w = -12...㉡ \\ w = 2...㉢ $$

이렇게 되면 상삼각 행렬이 만들어지고 가장 아래에 있는 행이 w=2 이기 때문에 나머지 값을 구할 수 있다. 이걸 행렬로 표현하면 다음과 같이 표현 할 수 있다. 이처럼 상삼각행렬이 만들어 졌을 때 대각 행렬들을 pivot이라고 부르고 ㉠의 u의 계수2를 1st pivot ㉡의 v의 계수 -8을 2nd pivot ㉢의 w의 계수를 3rd pivot 이라고 한다. pivot의 조건은 위에서 알 수 있듯이 모든 pivot은 non-zero이고 모든 pivot이 존재해야한다.

위와 같은 방정식 ㉠,㉡,㉢가 있을때 AX=B 라는 행렬로 만들어주면 다음과 같다.

가우스 소거법을 용이하게 하기위해 A와 B로만 표현하면 다음과 같이 표현 할 수 있다.

위의 연립방정식은 가우스 소거법으로 쉽게 해를 구할 수 있지만 이처럼 되지 않는 경우가 있다. 예를들어 소거법을 할 때 중간에 pivot이 0이 될 수도 있고, 모든 pivot이 존재하지 않을 수 있다. 이런경우의 예를 다음에서 살펴보자.

#####  - pivot이 중간에 0이 나오는 경우

$$u+v+w = a ...㉠ \\ 2u+2v+5w = b ...㉡ \\ 4u+6v+8w = c ...㉢$$

위와 같은 방정식이 있을 때 가우스 소거법을 진행하면 $㉡-2\times㉠$, $㉢-4\times㉠$ 을 해준다.

$$u+v+w = a...㉠ \\ 3w = b-2a...㉡\\2v+4w=c-4a...㉢$$

그러면 위와같이 ㉡ 방정식의 v의 계수가 사라져서 상삼각행렬을 만들 수 없다. 따라서 ㉡방정식과 ㉢방정식의 순서를 바꿔준다.


$$u+v+w = a...㉠ \\ 2v+4w=c-4a...㉡\\3w = b-2a...㉢$$

이렇게 소거법 중간에 pivot의 값이 0이 나올 때 순서를 바꿔주는것을 pivoting이라고 한다.

#####  - 모든 pivot이 존재하지 않을 때

$$u+v+w = a ...㉠ \\ 2u+2v+5w = b ...㉡ \\ 4u+4v+8w = c ...㉢$$

위와 같은 방정식이 있을 때 가우스 소거법을 진행하면 $㉡-2\times㉠$, $㉢-4\times㉠$ 을 해준다.

$$u+v+w = a...㉠ \\ 3w = b-2a...㉡\\4w=c-4a...㉢$$

그러면 위와 같이 ㉡,㉢의 식이 둘다 pivot이 0이 되기때문에 해가 무수히 많거나 해가 존재하지 않는다.

㉡과 ㉢ w에 대해 식을 정리하면 $$w = {b-2a\over3}, w={c-4a\over4} $$ 이다. 이때 $${b-2a\over3} ={c-4a\over4} $$ 이면 해가 무수히 많고 $${b-2a\over3} \neq {c-4a\over4} $$ 이면 해가 존재하지 않는다.

### Gauss Elimination code

```python
import numpy as np

def gauss(A):
    n=len(A)

    x= np.zeros(A.shape)
    temp =np.zeros(A.shape)

    x[0]= A[0]

    for i in range(n-1):
        for j in range(i+1,n):
            if A[i][i]== 0:
                temp[j] = A[j]
                A[j] = A[i]
                A[i] = temp[j]
                print(A)
            ratio = A[j][i]/A[i][i]
            x[j] = A[j]-ratio*A[i]
            A[j] = x[j]
```

```python
B = np.array([[1,1,1],
              [2,2,5],
              [4,6,8]])
```

```python
gauss(B)
```

### LU 분해

가우스 소거법의 과정을 행렬로 다음과 같이 행렬로 나타 낼 수 있다.













위처럼 $E_{21},E_{31},E_{32}$를 Elementary matrix라고 하고 Elementary matrix를 일반화 하면 다음과 같이 나타 낼 수있다.

$㉡ - l_{21}㉠=0$ 이되는 $l_{21}$을 원소를 가지는 행렬이 $E_{21}$이다

즉



마찬가지로 $E_{31},E_{32}$도 다음과 같이 표현 할 수 있다.

$ E_{32}E_{31}E_{21}A $ 에서 나온 값은 상삼각행렬(Upper triangular matrix)라고하고 U라고 한다.




elementary matrix는 항상 역행렬이 존재하므로 좌변과 우변에 $E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$를 곱해준다.

이러한 형태를 general case라고 하고 $E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$는 다음과 같다.

$E_{21}^{-1}E_{31}^{-1}E_{32}^{-1}$는 위와같이 항상 하삼각행렬(Lower triangular Matrix)이 나오고 L이라고 한다. 다시쓰면

$$A = LU$$

와 같이 나타 낼수 있고 이렇게 Elementary matrix로 표현하여 A를 하삼각행렬과 상삼각행렬의 곱으로 나타내주는 것을 LU 분해(LU factorization)이라고 한다. 이때 U를 pivot의 대각행렬과 나머지의 곱으로 나타낼 수 있는데 이를 LDU 분해라고 한다. 여기서 D는 pivot들의 diagnal matrix 라고 한다.

여기서 LDU, LU모두 A의 유일한 분해값이다.

앞에서, LDU, LU 모두 pivoting이 필요한 경우를 다루지 않았다. pivoting이 필요할 때의 방법을 알아보자. 

pivoting을 행렬로 표시하면 identity matrix의 pivoting이 필요한 해당 행을 바꿔주기만 하면된다.

여기서 $P^{-1} = P^T$이기 때문에 위 식을 일반화 하면 $$A = P^TLU$$이다.

### LU분해 code

```python
def LU(mat):
    n= len(mat)
    
    a=0

    for i in range(n):
        a = i+a

    E = np.zeros((a,n,n))
    L = np.zeros(mat.shape)
    P = np.zeros(mat.shape)
    temp = np.zeros(mat.shape)

    for i in range(a):
        for j in range(n):
            E[i][j][j] = 1
            L[j][j]=1
            P[j][j]=1

    cnt =0
    dot_last_cnt =0
    dot_first_cnt =0

    for i in range(n-1):
        for j in range(i+1,n):
            if mat[i][i] == 0:
                temp[j] = P[j]
                P[j] = P[i]
                P[i] = temp[j]
                mat = np.dot(P,mat)
            ratio=mat[j][i]/mat[i][i]
            E[cnt][j][i]=-ratio
            L[j][i]=ratio
            cnt = cnt+1
            dot_last_cnt = dot_last_cnt+1
        for p in range(dot_first_cnt,dot_last_cnt):
            mat = np.dot(E[p],mat)
        dot_first_cnt = dot_last_cnt

    U=mat

    print("A")
    print(np.dot(L,U))
    print("========================")
    print("L")
    print(L)
    print("========================")
    print("U")
    print(U)
```

```python
B = np.array([[1,1,1],
                [2,2,5],
                [4,6,8]])
```

```python
LU(B)
```

 ### Inverse and Transpose

행렬 A에 대하여 역행렬이 존재할때 행렬 A의 성질은 다음을 만족한다.

- Det(A) $\neq$ 0

- $Ax = b$ -> $A^{-1}b= x$ -> $AA^{-1} = A^{-1}A = I$ 

- $n \times n$ 행렬 A의 역행렬이 존재하면 n개의 pivot이 존재한다.

- 역행렬은 unique하다 즉 $A^{-1} = B\ or\ C$ 일 때, $(BA)C = B(AC) 이고 <=> C=B$ 이다.

- $A^{-1}$가 존재하면 $x$가 유일하다. 예를들어 $Ax=b -> x = A^{-1}b$ $A^{-1}$이 유일하기 때문에 x도 유일하다.

- non-zero vector $x$가 $Ax=0$을 만족하면 $A^{-1}$은 존재하지 않는다.

### Gauss-Jordan Method

역행렬을 구하는 Gauss-Jordan Method를 알아보자.

예를들어, 역행렬이 존재하는 행렬 A에 대해서 즉, $AA^{-1}=I$ 만족할때, $A^{-1}=[x_1,x_2,...,x_n]$ 이라하면 



$$AA^{-1}=A[X_1  X_2  ... X_n] = [AX_1 AX_2 ... AX_n] = I$$



위와 같이 열벡터로 나타날수 있기 때문에 가우스 소거법을 사용할 수 있다.

```python
import numpy as np

def gauss_jordan(A):
    n=len(A)

    x= np.zeros(A.shape)
    y= np.zeros(A.shape)
    temp =np.zeros(A.shape)
    A_inverse = np.identity(n)
    
    x[0]= A[0]

    for i in range(n-1):
        for j in range(i+1,n):
            if A[i][i]== 0:
                temp[j] = A[j]
                A[j] = A[i]
                A[i] = temp[j]
                print(A)
            ratio = A[j][i]/A[i][i]
            x[j] = A[j]-ratio*A[i]
            A[j] = x[j]
            y[j] = A_inverse[j]-ratio*A_inverse[i]
            A_inverse[j] =y[j]
    
    for k in reversed(range(n)):
        for l in range(k):
            ratio = A[l][k]/A[k][k]
            x[l] = A[l]-ratio*A[k]
            A[l] = x[l]
            y[l] = A_inverse[l]-ratio*A_inverse[k]
            A_inverse[l] =y[l]
    
    for p in range(n):
        ratio = A[p][p]
        A[p]=A[p]/ratio
        A_inverse[p]=A_inverse[p]/ratio
    
    print("A")
    print(A)
    print("========================")
    print("A_inverse")
    print(A_inverse)
    print("========================")
```

```python
B = np.array([[2,1,1],
                [4,-6,0],
                [-2,7,2]])
```

```python
gauss_jordan(B)
```
